```python
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def linear(x):
    return x
```

```python
def relu(x):
    return max(0,x)
```

```python
def tanh(x):
    return 1/(2*np.exp(-x))
```

```python
def perceptron(x1,x2):
    w1=0.5
    w2=0.5
    b=1
    y1=(w1*x1) + (w2*x2)+b
    y=sigmoid(y1)
    return y
```

```python
perceptron(10,6)
```

```python
import matplotlib.pyplot as plt

x=np.linspace(-10,10,400)

linear =x 
sigmoid=1/(1+np.exp(-x))
relu=np.maximum(0,x)
tanh=np.tanh(x)
```

```python

plt.plot(x, relu)
plt.title("ReLU")
plt.grid(True)
```
